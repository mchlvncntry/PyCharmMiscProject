```python
count=0
reps=10_000
die=np.arange(1,7)

for i in np.arange(reps):
    random_die_roll=np.random.choice(die)
    if random_die_roll >= 4:
        count += 1
prob_4_or_higher = count/reps
prob_4_or_higher
```
