```python
def flip_coin_10_times():
    num_heads = sum(np.random.choice([1,0], 10))
    return abs(num_heads-5)

print(flip_coin_10_times())
```

```python
#functional method, no need to make a list in memory like list comprehension
for __ in map(print, (flip_coin_10_times() for __ in np.arange(10))): pass
```

Complete the code below to simulate the experiment 10000 times and record the statistic in each of those trials in an array called abs_differences.

```python
trials = np.arange(10_000)
abs_differences = make_array()

for __ in trials:
    abs_diff_one_trial = flip_coin_10_times()
    abs_differences = np.append(abs_differences, abs_diff_one_trial)
```
